```python
import numpy as np
```

```python
X = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def predict(x,weight,bias):
    return weight*x + bias
```

```python
def cost_function(x,y,weight,bias):
    N = len(x)
    error = 0
    for i in range(N):
        error += (y[i]-(weight*x[i]+bias))**2
    return error/N
```

```python
def gradient_descent(x,y,weight,bias,learning_rate):
    dw=0
    db=0
    N=len(X)
    for i in range(N):
        dw += -2*x[i] * (y[i] - (weight*x[i] + bias))
        db += -2 * (y[i] - (weight*x[i] + bias))
    
    weight -= (dw / N) * learning_rate
    bias -= (db / N) * learning_rate
    
    return weight,bias
```

```python
def train(x,y,weight,bias,learning_rate,iters):
    cost_history=[]
    
    for i in range(iters):
        weight,bias = gradient_descent(x,y,weight,bias,learning_rate)
        
        cost = cost_function(x,y,weight,bias)
        cost_history.append(cost)
        
        if i%10==0:
            print("iter={}    weight={}    bias={}    cost={}".format(i, weight, bias, cost))
            
        if i>2:
            if (cost_history[i-1]-cost_history[i])<0.0000000001:
                print(i)
                break
    return weight,bias
```
